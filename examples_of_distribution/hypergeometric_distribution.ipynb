**Definition** Consider a box with *w* white balls and $b$ black balls. We draw $n$ balls out of the box at random without replacement, such that all ${w+b \choose n}$ samples are equally likely. Let $X$ be the number of white balls in the sample. Then $X$ is said to have the *hypergeometric distribution* with parameters $w,b,$ and $n$. We write $X \sim \text{HGeom}(w,b;n)$  

The probability mass function of $X$ is given by 

$$P(X=k) = \frac{ {w \choose k}{b \choose n-k}}{w+b \choose n}$$

for integers $k$ satisfying $0 \leq k \leq w$ and $0 \leq n-k \leq b$, and $0$ otherwise.

**Expected value and variance**\
Let $X \sim \text{HGeom}(w,b;n)$, then $E(X) = n \frac{w}{w+b}$ and $\text{Var}(X) = n  \frac{w}{w+b} \frac{b}{w+b} \frac{w+b-n}{w+b-1} $.

```python
from scipy.stats import hypergeom
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set(rc={'figure.figsize':(5,6)})

#here N = w+b
[N, w, n] = [25, 8, 10]

data_hypergeom = hypergeom.rvs(N, w, n,size=10000)
ax = sns.distplot(data_hypergeom,
                  kde=False,
                  color='b',
                  hist_kws={'alpha':1})
ax.set(ylabel='count', title = 'sampling from a hypergeometric distribution')
pass
```

```python
[N, w, n] = [25, 8, 10]

rv = hypergeom(N, w, n)
x = np.arange(0, w+1)

plt.plot(x, rv.pmf(x), 'bo')
plt.vlines(x, 0, rv.pmf(x), lw=2)
plt.ylabel('PMF')
plt.show()
```
