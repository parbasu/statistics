**Definiton** Let $$T = \frac{Z}{\sqrt{V/n}},$$ where $Z$ is the standard normal random variable, $V$ is the chi-square distribution with $n$ degrees of freedom, and $Z$ is independent of $V$. Then $T$ is said to have the *Student-$t$ distribution with $n$ degrees of freedom*. We write $T \sim t_n$. Often *Student $t$-distribution* is abbreviated to *$t$-distribution*.

**Expected value and variance**\
Let $T \sim t_n$, then $E(T) = 0$ and $\text{Var}(T)= \frac{n}{n-2}$ for $n > 2$ and $\infty$ for $n=2$.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#plotting the PDF

sns.set()

x = np.arange(-5,5,0.05)

plt.plot(x, stats.t.pdf(x, df =2))
plt.xlim(-5,5)
plt.ylabel('PDF')
plt.title('Student-t distribution with  2  degrees of freedom')
pass
```

```python
#plotting the CDF

x = np.arange(-5,5,0.05)

plt.plot(x, stats.t.cdf(x, df =2))
plt.xlim(-5,5)
plt.ylabel('CDF')
plt.title('Student-t distribution with  2  degrees of freedom')
pass
```

```python
#the student-t distribution with 2 degrees of freedom has fatter tails

sns.set(rc={'figure.figsize':(6,6)})

x = np.arange(-5,5,0.05)

plt.plot(x, stats.t.pdf(x, df =2), label='student-t with df = 2')
plt.plot(x, stats.norm.pdf(x), label = 'standard normal')
plt.xlim(-5,5)
plt.ylabel('PDF')
plt.legend(loc=1)
pass

```

```python
#plotting the PDF for different degrees of freedom

x = np.arange(-5,5,0.05)

plt.plot(x, stats.t.pdf(x, df =2), label='df = 2')
plt.plot(x, stats.t.pdf(x, df =4), label='df = 4')
plt.plot(x, stats.t.pdf(x, df =6),label='df = 6')
plt.xlim(-5,5)
plt.legend()
plt.ylabel('PDF')
plt.title('Student-t distribution with  2  degrees of freedom')
pass
```

```python
#plotting the CDF for different degrees of freedom

x = np.arange(-5,5,0.05)

plt.plot(x, stats.t.cdf(x, df =2), label='df = 2')
plt.plot(x, stats.t.cdf(x, df =4), label='df = 4')
plt.plot(x, stats.t.cdf(x, df =6),label='df = 6')
plt.xlim(-5,5)
plt.legend()
plt.ylabel('CDF')
plt.title('Student-t distribution with  2  degrees of freedom')
pass
```

```python
#sampling from a student-t distribution

data_t = stats.t.rvs(size=10000,loc=0,scale=1, df=5)
ax = sns.distplot(data_t, kde=False, bins = 100,
                 color= 'b',
                 hist_kws={'alpha': 1})
ax.set(ylabel='count', title = 'sampling from a student-t distribution with df=5') #df - degrees of freedom
pass
```
