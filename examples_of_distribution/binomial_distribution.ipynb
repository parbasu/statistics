**Definition** Suppose that $n$ independent Bernoulli trials are performed, each with the same sucess probability $p$. Let $X$ be the number of successes. The distribution of $X$ is called the *Binomial distribution* with paramaters $n$ and $p$. We write $X \sim \text{Bin}(n,p)$.

The probability mass function of $X$ is given by $$P(X=k) = {n \choose k} p^k (1-p)^{n-k}$$
for $k = 0,1,\dots,n$ and it is equal to $0$ otherwise.

```python
from scipy.stats import binom
import seaborn as sns

sns.set(rc={'figure.figsize':(5,6)})

data_binom = binom.rvs(n=10,p=0.5,size=10000)
ax = sns.distplot(data_binom,
                  kde=False,
                  color='b',
                  hist_kws={'alpha':1})
ax.set(ylabel='count', title = 'sampling from a Binomial distribution')
pass
```

```python
import matplotlib.pyplot as plt
import numpy as np

sns.set()

n=10
p=0.5
x = np.arange(n+1)

plt.plot(x, binom.pmf(x, n, p), 'o--')
plt.ylabel('PMF')
plt.title('binomial distribution')
pass
```

```python
#we plot three binomial distribution with different parameters

ns= [20,20,40]
ps = [0.5,0.7,0.5]

for (p,n) in zip(ps,ns):
    bd = binom(n,p)
    x = np.arange(n+1)
    plt.plot(x, bd.pmf(x), 'o--', label = 'p ={0:3.1f}, n ={1}'.format(p,n))
    
plt.legend()
plt.ylabel('PMF')
plt.title('binomial distributions')
pass
```
