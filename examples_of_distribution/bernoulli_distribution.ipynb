**Definition** A random variable $X$ is said to have the *Bernoulli distribution* with parameter $p$ if $P(X=1) = p$ and $P(X=0) = 1-p$, where $0 < p < 1$. We write this as $X \sim \text{Bern}(p)$

**Definition** The *indicator random variable* of an event $A$ is the random variable which equals $1$ if $A$ occurs and $0$ otherwise. We denote it by $I_A$. Note $ I_A \sim \text{Bern}(p)$, where $p = P(A)$.

**Definition** An experiment that can result in either a *success* or a *failure* (but not both) is called a *Bernoulli trial*. A Bernoulli random variable can be thought of as *the indicator of success* in a Bernoulli trial: it equals $1$ if success occurs and $0$ if failure occurs in the trial.

```python
from scipy.stats import bernoulli
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Generating random numbers from Bernoulli distribution

sns.set(rc={'figure.figsize':(5,6)})

data_bern = bernoulli.rvs(size=10000,p=0.3)
ax = sns.distplot(data_bern, kde=False,
                 color= 'b',
                 hist_kws={'alpha': 1})
ax.set(ylabel='count', title = 'sampling from a Bernoulli distribution')
pass
```

```python
#Note that this is normalized in the sense that the total area under the histogram is equal to 1.

ax = sns.distplot(data_bern, kde=False,
                 color= 'b', norm_hist = True,
                 hist_kws={'alpha': 1})
ax.set(title = 'normalized histogram')
pass
```

```python
plt.plot([0,1], bernoulli.pmf([0,1], 0.5), 'o')
plt.ylabel('PMF')
plt.title('Bernoulli distribution')
pass
```
