**Definition** Let $V = Z_1^2 + \dots + Z_n^2$, where $Z_1, \dots, Z_n$ are independent and identically distributed standard normal random variables. Then $V$ is said to have the *Chi-square distribution with $n$ degrees of freedom*. We write this as $V \sim \chi_{n}^{2}$.

**Expected value and variance**\
Let $V \sim \chi_{n}^{2}$, then $E(V) = k$ and $\text{Var}(V)=2k$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
sns.set()

x = np.arange(0,8,0.05)

plt.plot(x, stats.chi2.pdf(x, 2))
plt.xlim(0,8)
plt.ylabel('PDF')
plt.title('chi-square distribution with 2 degree of freedom')
pass
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

x = np.arange(0,8,0.05)

plt.plot(x, stats.chi2.cdf(x, 2))
plt.xlim(0,8)
plt.ylabel('CDF')
plt.title('chi-square distribution with 2 degree of freedom')
pass
```

```python
#plotting PDF for different degrees of freedom

x = np.arange(0,8,0.05)
degree_of_freedoms = [1,2,3,5]

for degree in degree_of_freedoms:
    plt.plot(x, stats.chi2.pdf(x, degree), label = 'k={0}'.format(degree))

plt.legend()
plt.xlim(0,8)
plt.ylabel('PDF')
pass
```

```python
#plotting chi-square distributions with different degrees of freedom

x = np.arange(0,8,0.05)
degree_of_freedoms = [1,2,3,5]

for degree in degree_of_freedoms:
    plt.plot(x, stats.chi2.cdf(x, degree), label = 'k={0}'.format(degree))

plt.legend()
plt.xlim(0,8)
plt.ylabel('CDF')
pass
```

```python
#sampling from a chi-square distribution

import seaborn as sns


data_chi2 = stats.chi2.rvs(size=10000,loc=0,scale=1, df=5)
ax = sns.distplot(data_chi2, kde=False, bins = 100,
                 color= 'b',
                 hist_kws={'alpha': 1})
ax.set(ylabel='count', title = 'sampling from a chi-square distribution with df=5') 
```
