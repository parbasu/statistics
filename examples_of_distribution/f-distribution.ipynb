**Definition** Let $$X = \frac{U_1/d_1}{U_2/d_2}$$ where
where $U_1$ and $U_2$ are indenpedent chi-square distributions with $d_1$ and $d_2$ degrees of freedom. Then $X$ is said to have the *$F$-distribution* with parameters $d_1$ and $d_2$.

The expected value and variance of a $F$-distribution are a bit complicated, if you are curious google it!

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#plotting the PDF

sns.set()

x = np.arange(0,3,0.01)

plt.plot(x, stats.f.pdf(x, 2,1))
plt.xlim(0,3)
plt.ylabel('PDF')
plt.title('F-distribution with parameters (2,1)')
pass
```

```python
#plotting the CDF

sns.set()

x = np.arange(0,3,0.01)

plt.plot(x, stats.f.cdf(x, 2,1))
plt.xlim(0,3)
plt.ylabel('CDF')
plt.title('F-distribution with parameters (2,1)')
pass
```

```python
#plotting the PDF for different parameters

x = np.arange(0,3,0.01)

d1s = [1,2,6,100]
d2s = [1,1,9,100]

for (d1,d2) in zip(d1s, d2s):
    plt.plot(x, stats.f.pdf(x, d1,d2), label = '{0},{1}'.format(d1,d2))

plt.xlim(0,3)
plt.ylabel('PDF')
plt.legend()
plt.title('F-distribution with parameters (d1,d2)')
pass
```

```python
#plotting the CDF for different parameters

x = np.arange(0,3,0.01)

d1s = [1,2,6,100]
d2s = [1,1,9,100]

for (d1,d2) in zip(d1s, d2s):
    plt.plot(x, stats.f.cdf(x, d1,d2), label = '{0},{1}'.format(d1,d2))

plt.xlim(0,3)
plt.ylabel('CDF')
plt.legend()
plt.title('F-distribution with parameters (d1,d2)')
pass
```

```python
#sampling from a F-distribution

data_f = stats.f.rvs(6,9, size=10000,loc=0,scale=1)
ax = sns.distplot(data_f, kde=False, bins = 100,
                 color= 'b',
                 hist_kws={'alpha': 1})
ax.set(ylabel='count', title = 'sampling from a F-distribution with parameters (6,9)')
pass
```
