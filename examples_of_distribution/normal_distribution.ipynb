**Definition** A continuous random variable $Z$ is said to have the *standard Normal distribution* if its probability density function $\phi$ is given by $$\phi(z) = \frac{1}{\sqrt{2\pi}} e^{-z^2/2}, \ -\infty < z < \infty.$$
We write $Z \sim \mathcal{N}(0,1)$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

z = np.arange(-5,5,0.05)
normal = norm.pdf(z)
plt.plot(z, normal, '-', color ='b')
plt.ylabel('PDF')
plt.title('normal distribution')
pass
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns

sns.set()

z = np.arange(-5,5,0.05)
normal = norm.cdf(z)
plt.plot(z, normal, '-', color ='b')
plt.ylabel('CDF')
plt.title('normal distribution')
pass
```

```python
#plotting normal distribution with three different parameters

from scipy import stats

mean = [0,0,0,-2]
standard_deviation = [0.2,1,5,0.2]
z = np.arange(-5,5,0.02)

sns.set(rc={'figure.figsize':(6,4)})

with sns.color_palette('hls', 4):
    for mu,sigma in zip(mean, np.sqrt(standard_deviation)):
        y = stats.norm.cdf(z, mu, sigma)
        plt.plot(z,y, label = '$\mu = {0}, \; \t\sigma={1:3.1f}$'.format(mu,sigma)) # format string

plt.legend()
plt.ylabel('CDF')
plt.title('normal distribution')
pass
```

```python
#sampling from a normal distribution

import seaborn as sns

sns.set(rc={'figure.figsize':(5,6)})

data_normal = norm.rvs(size=10000,loc=0,scale=1)
ax = sns.distplot(data_normal, kde=False, bins = 100,
                 color= 'b',
                 hist_kws={'alpha': 1})
ax.set(ylabel='count', title = 'sampling from a normal distribution')
pass
```

```python
#empirical rule

sns.set(rc={'figure.figsize':(6,6)})

z = np.arange(-5,5,0.05)
normal = norm.pdf(z)
plt.plot(z, normal, '-', color ='b')
plt.ylabel('PDF')
plt.title('normal distribution')
plt.vlines(1, 0, norm.pdf(1), color = 'g', label = '68%')
plt.vlines(-1, 0, norm.pdf(-1), color = 'g')

plt.vlines(2, 0, norm.pdf(2), color = 'y', label = '95%')
plt.vlines(-2, 0, norm.pdf(-2), color = 'y')

plt.vlines(3, 0, norm.pdf(3), color = 'r', label = '99.7%')
plt.vlines(-3, 0, norm.pdf(-3), color = 'r')

plt.ylim(ymin=0)
plt.legend()
pass
```
