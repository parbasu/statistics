**Definition** A continuous random variable $X$ is said to have the *Exponential distribution* with parameter $\lambda$, where $\lambda > 0$, if its PDF is $$f(x) = \lambda e^{-\lambda x}, \ x > 0.$$

```python
#plotting exponential distribution with three different parameters
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

x = np.arange(0,3,0.01)

plt.plot(x, stats.expon.pdf(x,0, 1))
plt.xlim(0,3)
plt.ylabel('PDF')
plt.title('exponential distribution')
pass
```

```python
#plotting exponential distribution with three different parameters
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

x = np.arange(0,3,0.01)

plt.plot(x, stats.expon.cdf(x,0, 1))
plt.xlim(0,3)
plt.ylabel('CDF')
plt.title('exponential distribution')
pass
```

```python
#plotting exponential distribution with three different parameters

x = np.arange(0,3,0.01)
lambdas = [0.5,1,1.5]

for par in lambdas:
    plt.plot(x, stats.expon.pdf(x,0, par), label = '$\lambda = {0:3.1f}$'.format(par))
             
plt.xlim(0,3)
plt.legend()
pass
```

```python
#plotting exponential distribution with three different parameters

x = np.arange(0,3,0.01)
lambdas = [0.5,1,1.5]

for par in lambdas:
    plt.plot(x, stats.expon.cdf(x,0, par), label = '$\lambda = {0:3.1f}$'.format(par))
             
plt.xlim(0,3)
plt.legend()
pass
```

```python
#sampling from an exponential distribution

import seaborn as sns

sns.set(rc={'figure.figsize':(5,6)})

data_expon = stats.expon.rvs(size=10000,loc=0,scale=1)
ax = sns.distplot(data_expon, kde=False, bins = 100,
                 color= 'b',
                 hist_kws={'alpha': 1})
ax.set(ylabel='count', title = 'sampling from an exponential distribution')
pass
```
