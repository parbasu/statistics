**Poisson distribution** A random variable $X$ has the *Poisson distribution* with parameter $\lambda$, where $\lambda >0$, if the PMF of $X$ is $$P(X=k) = \frac{e^{-\lambda} \lambda^k}{k!}, \ k = 0,1,2,\dots$$
We write $X \sim Pois(\lambda)$.

```python
from scipy.stats import poisson
import seaborn as sns

sns.set(rc={'figure.figsize':(5,6)})

data_poisson = poisson.rvs(mu = 3, size=10000) # can't use lambda here 
ax = sns.distplot(data_poisson,
                  kde=False,
                  color='b',
                  hist_kws={'alpha':1})
ax.set(ylabel='count', title = 'sampling from a Poisson distribution')
pass
```

```python
import matplotlib.pyplot as plt
import numpy as np

sns.set()


k=np.arange(20)
plt.plot(k, poisson.pmf(k, 4), 'o--')

plt.ylabel('PMF')
plt.title('poisson distribution')
pass
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
import seaborn as sns

sns.set(rc={'figure.figsize':(6,7)})

X = np.arange(0,21)

col = {1: 'orange', 4: 'purple', 10: 'lightblue'}


A = []
for L in 1,4,10:

    P = poisson.cdf(X,L)

    for k in range(1,len(X)):
        plt.plot([k-1,k], [P[k-1],P[k-1]], '-', color='grey')

    a = plt.plot(X, P, 'o', color=col[L], label = '$\lambda$={0}'.format(L))
    A.append(a)
    
plt.ylabel('CDF')
plt.legend()

plt.xlim(-1,21)

pass
```

```python
import matplotlib.pyplot as plt
import numpy as np

sns.set()

lambdas = [1,4,10]
k=np.arange(20)

for par in lambdas:
    plt.plot(k, poisson.pmf(k, par), 'o--', label='$\lambda={0}$'.format(par))

plt.ylabel('PMF')
plt.legend()
plt.title('poisson distributions')
pass
```
