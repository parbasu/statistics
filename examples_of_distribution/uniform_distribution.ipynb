**Definition** A continous random variable $X$ is said to have the *uniform distribution* on the interval $(a,b)$ if its probability density function is 

$$f(x) = \begin{cases}
\frac{1}{b-a}  & \text{if $a < x < b$} \\
0 & \text{otherwise.}
\end{cases} $$ 

```python
#plotting the pdf

from scipy.stats import uniform
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set()

#breaking the graph into three parts
x_1 = np.arange(-1,0,0.05)
x_2= np.arange(0.05,10,0.05)
x_3=np.arange(10.05,11,0.05)

#plotting the pmf corresponding to the three parts above
plt.plot(x_1, uniform.pdf(x_1, loc=0, scale=10), color='b')
plt.plot(x_2, uniform.pdf(x_2, loc=0, scale=10),color = 'b')
plt.plot(x_3, uniform.pdf(x_3, loc=0, scale=10),color = 'b')

#adding vertical lines at jump discontinuities 
plt.vlines(0,0,0.10,linestyle='dotted',color = 'b')
plt.vlines(10,0,0.10,linestyle='dotted',color = 'b')

#adding markers at jump discontinuities
plt.plot(0,0, marker = 'o' , markerfacecolor= 'none' , markersize =4)
plt.plot(10,0, marker = 'o' , markerfacecolor= 'none', color='b', markersize =4)
plt.plot(0,0.10, marker = 'o', color='b' , markersize =4)
plt.plot(10,0.10, marker = 'o', color='b', markersize =4)

plt.ylabel('PDF')
plt.title('uniform distribution')

pass
```

```python
#plotting the cdf

x = np.arange(-2,12,0.05)
plt.plot(x, uniform.cdf(x, loc=0,scale=10))

plt.title('uniform distribution')
plt.ylabel('CDF')

pass
```

```python
#sample of size 10

from scipy.stats import uniform
import seaborn as sns

sns.set(rc={'figure.figsize':(5,6)})

data_uniform = uniform.rvs(size=10, loc = 0, scale=10)   
ax = sns.distplot(data_uniform,
                  kde=False, norm_hist = True,
                  color='b', 
                  hist_kws={'alpha':1})
ax.set(ylabel='density', title = 'sampling from a uniform distribution')
pass
```
